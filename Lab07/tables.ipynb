```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_asym = pd.read_csv('phil_asym.csv')
df_conductor = pd.read_csv('phil_conductor.csv')

plt.bar(df_conductor['Philosophers Number'].unique() - 0.2, df_conductor['Time'].groupby(df_conductor['Philosophers Number']).mean(), width=0.4, label='Rozwiązanie z kelnerem')
plt.bar(df_asym['Philosophers Number'].unique() + 0.2, df_asym['Time'].groupby(df_asym['Philosophers Number']).mean(), width=0.4, label='Rozwiązanie asymetryczne')
plt.title('Porównanie rozwiązań problemu pięciu filozofów')
plt.xlabel('Liczba filozofów')
plt.ylabel('Średni czas oczekiwania')
plt.legend()
plt.show()
```
